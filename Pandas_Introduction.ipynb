```python
# Apply a function to the "Enrollment" column. 
# Note: the Enrollment column is a string, so we need to convert it to integer before doing a calculation.

df['Enrollment'] = df['Enrollment'].astype(int)  # Convert "Enrollment" to int
df['Enrollment_in_thousands'] = df['Enrollment'].apply(lambda x: x // 1000)  # Integer division by 1000
df['Enrollment_in_thousands'] = df['Enrollment_in_thousands'].astype(str) + 'k'  # Convert back to string and append 'k'
```

```python
df.head(5)
```

```python
# Find the number of institutions per continent
df['Continent'].value_counts()
```

```python
# Find the average enrollment (in thousands) per continent
df['Enrollment_in_thousands'] = df['Enrollment_in_thousands'].str.replace('k', '').astype(int)  # remove 'k' and convert to int
df.groupby('Continent')['Enrollment_in_thousands'].mean()
```

```python
# Average 'Enrollment' per 'Continent'
avg_enrollment_per_continent = df.groupby('Continent')['Enrollment'].mean()

# Filter continents where the average 'Enrollment' is greater than 20000
avg_enrollment_per_continent[avg_enrollment_per_continent > 20000]
```

```python
# Count the number of universities per 'Continent' and 'Affiliation'
universities_per_continent_and_affiliation = df.groupby(['Continent', 'Affiliation']).size()

# Filter groups where the number of universities is greater than 10
universities_per_continent_and_affiliation[universities_per_continent_and_affiliation > 10]
```

```python
# Filter 'Continent' groups having more than 5 universities with 'Enrollment' greater than 20000
df[df['Enrollment'] > 20000].groupby('Continent').filter(lambda x: len(x) > 5)
```
